```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
PATH = os.path.join('c:' + os.sep, 'Temp2', 'data.csv')
df = pd.read_csv(PATH)
print(df)
```

```python
%matplotlib inline
plt.title('Find Dynamic pricing',fontsize=24)
plt.xlabel('Area (Sq feet)',fontsize=14)
plt.ylabel('Price ( US $)',fontsize=14)
plt.scatter(df.area, df.price,color ='red' , marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
print(reg)
```
